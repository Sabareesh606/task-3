```python
column=df[['age','campaign','duration']]
q1=np.percentile(column,25)
q3=np.percentile(column,75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
df[['age','campaign','duration']]=column[(column>lower_bound) & (column < upper_bound)]
```

```python
#to exclude non numeri values
n_df=df.drop(columns=c_cols)
#to compute correlations
corr=n_df.corr()
print(corr)
#filtering correlations which are >=0.90
corr=corr[abs(corr)>=0.90]
sns.heatmap(corr,annot=True,cmap='Set2',linewidth=0.2)
plt.show()
```

```python
high_corr_cols=['emp.var.rate','euribor3m','nr.employed']
```

```python
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
#tranforming characters into numericals
lb=LabelEncoder()
df_encoded=df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
x=df_encoded.drop('deposit',axis=1)
y=df_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
#to remove unwanted rows and columns
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print("Accuracy_Score",acc)
    cm=confusion_matrix(y_test,y_pred)
    print("\nConfusion Matrix\n",cm)
    print("Classification Report",classification_report(y_test,y_pred))
def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    print('Trainig Score',train_score)
    print('Testing Score',test_score)
                                                   
```

```python
dt=DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```
